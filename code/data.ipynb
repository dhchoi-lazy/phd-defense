```python
import pandas as pd
import re
```

```python
bangmok = pd.read_csv('bangmok.csv')
```

```python
cnt = pd.DataFrame(bangmok.groupby('pass_year').size()).reset_index().rename(columns={0:'count', 'pass_year':'year'})
# cnt.to_json('bangmok_year.json', orient='records')
```

```python
freq = bangmok[bangmok['rank']==1].groupby(['pass_year','rank']).size().reset_index()
freq.columns = ['year','rank','freq']
# freq[['year','freq']].to_json('bangmok_year_freq.json', orient='records')
```

```python
pd.merge(cnt, freq, on='year')[['year','count','freq']].to_json('bangmok_count.json', orient='records')
```

```python
bm
```

```python
bm['pass_age'] = bm['pass_year'] - bm['birth']
bm['lifespan'] = bm['death'] - bm['birth']
bm['office_span'] = bm['death'] - bm['pass_year']
```

```python
import requests
from lxml import html

d = html.fromstring(requests.get("https://ko.wikipedia.org/wiki/%EC%A1%B0%EC%84%A0_%EA%B5%AD%EC%99%95").text).xpath('//*[@id="mw-content-text"]/div[1]/center[1]/table/tbody//tr/td[3]')
```

```python
kings = [ i.text.strip('\n').split(' - ') for i in d]
kings = [[int(k[0].replace('년', "")), int(k[1].replace('년', ""))] for k in kings]
```

```python
kings = pd.DataFrame(kings, columns=['birth','death'])
```

```python
king_name = "태조 정종 태종 세조 문종 단종 세조 예종 성종 연산 인종 명종 선조 광해 인조 효종 현종 숙종 경종 영조 정조 순조 헌종 철종 고종".split(' ')
kings['king'] = pd.Series(king_name)
```

```python
kings['lifespan'] = kings['death'] - kings['birth']
```

```python
kings.to_csv('kings.csv', index=False)
```

```python
import json
with open('gwageo.json', 'r') as f:
    gwageo = json.load(f)
gwageo
```

```python
import pandas as pd
gnodes = pd.DataFrame(gwageo['nodes'])
```

```python
gnodes
```

```python
glinks = pd.DataFrame(gwageo['links'])
```

```python
nd = {k:v for k, v in zip(gnodes.node, gnodes.name)}
```

```python
glinks['source_name'] = glinks['source'].map(nd)
glinks['target_name'] = glinks['target'].map(nd)
glinks
```

```python
glinks.to_csv('gwageo_links.csv', index=False)
```

```python
glinks['source_name2'] = glinks['source_name'].apply(lambda x: x.split(" - ")[1])
glinks['source_target2'] = glinks['target_name'].apply(lambda x: x.split(" - ")[1])
```
